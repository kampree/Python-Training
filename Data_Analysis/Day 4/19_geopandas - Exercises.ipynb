When we first learned pandas, we learned about the `groupby` operation. geopandas provides a similar function called `dissolve`. It is basically the same as groupby for the non-geometry column, but will combine all the geometries by overlapping them. 

One annoying thing about `dissolve` is that it operates on all columns, so we first have to make a smaller GeoDataFrame that only has the columns we want to group by or aggregate.

```python
#Select population estimates of every countries
populations = countries_wo_antarctic[['CONTINENT', 'POP_EST', 'geometry']]
```

```python
#Aggregate Sum by CONTINENT
population_by_continent = populations.dissolve(by='CONTINENT', aggfunc='sum')
ax = population_by_continent.plot(column='POP_EST', legend=True, figsize=(15,7))
ax.set_axis_off()
ax.set_title("Population estimates by continent")
plt.savefig('population_by_continent_map.png')
```
