# 12. When to use a groupby command in pandas?

Drinks dataset: https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv

Source: https://github.com/fivethirtyeight/data/tree/master/alcohol-consumption

```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv")
drinks.head()
```

```python
drinks[drinks['country']=='Thailand']
```

```python
drinks[drinks['continent']=='Asia']
```

What is the average beer servings across all countries?

```python
drinks.beer_servings.mean()
```

What is the average beer servings by continent?

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()['Africa']
```

```python
drinks[drinks['continent'] == 'Africa'].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

Answer: when want to analyse by some category (e.g. beer servings by continent)

```python
drinks.groupby('continent')
```

## Across all numeric columns

```python
by_continent = drinks.groupby('continent')
by_continent.mean()
```

```python
by_continent.std()
```

```python
by_continent.min()
```

```python
drinks[drinks['continent']=='South America']
```

```python
by_continent.max()
```

```python
by_continent.beer_servings.describe()
```

```python
by_continent.describe()
```

```python
by_continent.describe().transpose()
```

```python
by_continent.describe().transpose()['Asia']
```

## What to do with missing value?

No right or wrong answer

### 1. Drop Missing values

```python
ufo.shape
```

Drop a row if any of columns has a missing value

```python
ufo.dropna()
```

Drop a row if either city or shape reported are missing

```python
ufo.dropna(subset=["City", "Shape Reported"])
```

### 2. Filling missing values

By default, missing values are excluded in value_counts()

```python
ufo['Shape Reported'].value_counts()
```

```python
ufo['Shape Reported'].value_counts().sum()
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
```

```python
ufo['Shape Reported'].fillna(value='Unknown', inplace=True)
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
```
