# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Styles

You can set particular styles:

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

## Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

## Height and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: height, and aspect. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

## Grid Type Plot

Works with relplot, displot, catplot, lmplot, pairplot and jointplot only

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',height=4, aspect=4, data=tips)
```

```python
sns.relplot(x='tip',y='total_bill',data=tips, height=4, aspect=4)
```

```python
sns.displot(x='tip',data=tips, kind='hist', height=4, aspect=4)
```

## Scale and Context

The set_context() allows you to override default parameters:

```python
#Default
sns.set_context("notebook")
```

```python
sns.set_context('poster', font_scale=1.2)
sns.lmplot(x='total_bill',y='tip',height=7, aspect=2, data=tips)
```
